# Support Vector Machine

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC, LinearSVC
%matplotlib inline
plt.style.use('seaborn-white')
```

## Example

```python
df=pd.DataFrame({'temperature':[85,60,80,72,68,74,69,75,83],
                 'humidity':[85,70,90,95,80,73,70,85,78],
                 'play':['no','yes','no','no','yes','yes','yes','no','no'] })
df
```

```python
df['play']=df['play'].map({'no': -1, 'yes': 1})
```

```python
df
```

```python
# 5x+2y-535=0
plt.figure()
plt.scatter(x=df['temperature'],y=df['humidity'], c=df['play'].map({-1:'blue',1:'red'}))
xx=np.linspace(55,90,50)
y1=(530-5*xx)/2
y2=(520-4*xx)/3
y3=(995.5-7*xx)/6
plt.plot(xx,y1,c='k',linewidth=2,label='line 1')
plt.plot(xx,y2,c='g',linewidth=2,label='line 2')
plt.plot(xx,y3,c='red',linewidth=2,label='line 3')
plt.xlabel('temperature')
plt.ylabel('humidity')
plt.title('(1) 5x+2y-530=0, (2) 4x+3y-520=0, (3) 7x+6y-995.5=0')
plt.xlim(55,90)
plt.ylim(60,100)
plt.legend()
plt.show()
```

```python
# x+y-5=0
plt.figure(figsize=(6,6))
xx=np.linspace(1,5,50)
y1=(5-xx)
plt.plot(xx,y1,c='b',linewidth=2)
xx1=np.linspace(2,4,20)
y2=xx1-1   
plt.plot(xx1,y2,c='g',linewidth=2,linestyle='--')
plt.scatter(2,1,c='r',marker='o')
plt.scatter(4,3,c='b',marker='o')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('x+y-5=0')
plt.xlim(1,5)
plt.ylim(1,5)
plt.axis('equal')
plt.show()
```
